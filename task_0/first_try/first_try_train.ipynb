```python
import pandas as pd
df = pd.read_csv("first_try_data.csv", header=0)
# simple classifier
df.loc[df['y'] < 60, 'y'] = 0
df.loc[df['y'] >= 60, 'y'] = 1
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.preprocessing import StandardScaler

all_components = [
    'BigBlueButtonBN',
    'Fail',
    'Faili esitamine',
    'Foorum',
    'Hindajaraport',
    'Kaust',
    'Kommentaarid tööle',
    'Küsitlus',
    'Leht',
    'Logid',
    'Minu hinderaport',
    'Onlain teksti esitamine',
    'Osalemine',
    'Prügikast',
    'Rühma valimine',
    'Süsteem',
    'Tagasiside',
    'Tegevuste aruanne',
    'Test',
    'Tund',
    'URL',
    'Virtual programming lab',
    'Õppijaraport',
    'Ülesanne',
    'Ülevaate aruanne'
]

scaler = StandardScaler()
df[all_components + ['label']] = scaler.fit_transform(df[all_components + ['label']])
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=164654)
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()
forest.fit(train[all_components + ['label']], train['y'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

ftnames = df.columns[:-1]
results = []
y = []
i = 0
for imt in forest.feature_importances_:
    imt = round(imt,4)
    results.append((ftnames[i], imt))
    i = i + 1


results.sort(key=lambda importance: importance[1])
results.reverse()


features = ftnames
importances = forest.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(6, 6.5))
plt.barh(range(len(indices)), importances[indices], color='b', align='center')

for s in results:
    y.append(s[0])
    
y.reverse()
plt.yticks(np.arange(len(y)), y)
plt.xlabel('Relative Importance')
plt.show()
```

```python
y.reverse()
selected_feature = y[:19]
selected_feature
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(train[selected_feature],train['y'])
```

```python
y_pred = model.predict(test[selected_feature])
y_pred
```

```python
print(f"Accuracy {np.sum(y_pred == test['y']) / len(y_pred) * 100:.2f}%")
```

```python
print(f"Recall rate {(np.sum(y_pred == 0) / np.sum(test['y'] == 0)) * 100:.2f}%")
print(f"Recall rate of (grade < 60) students is not good at all")
```
