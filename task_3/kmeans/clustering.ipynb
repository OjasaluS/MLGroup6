```python
from typing import List

import pandas as pd
from pandas import DataFrame

RANDOM_STATE = 111

dataframe = pd.read_csv("data.csv", header=0)
dataframe = dataframe.set_index("student_id")
```

```python
dataframe
```

```python
dataframe.describe().T
```

```python
from sklearn.preprocessing import MinMaxScaler


def get_keys(df: DataFrame) -> List[str]:
    result = []

    for col in df.columns:
        result.append(col)

    return result


def get_scale_keys(df: DataFrame) -> List[str]:
    result = []

    for col in df.columns:
        result.append(col)

    return result


df_clust = dataframe.copy()

keys = get_keys(dataframe)
scale_keys = get_scale_keys(dataframe)

scaler = MinMaxScaler()
df_clust[scale_keys] = scaler.fit_transform(df_clust[scale_keys])
```

```python
print(df_clust.shape)
df_clust.head()
```

```python
df_clust.describe().T
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score as ss
from tqdm import tqdm

k1 = []
ss_core1 = []
inertia_s1 = []

for i in tqdm(range(2,21)):
    k1.append(i)
    kmeans1 = KMeans(
        n_clusters=i,
        random_state=RANDOM_STATE,
        max_iter=100
    ).fit(df_clust)
    ss_core1.append(ss(df_clust,kmeans1.labels_))
    inertia_s1.append(kmeans1.inertia_)

plt.figure(figsize=(15, 5))
plt.plot(
    k1,
    inertia_s1,
    color="blue",
    linestyle="dashed",
    marker="o",
    markerfacecolor="red",
    markersize=10
)
plt.title("Inertia (SSE) vs. K Value")
plt.xlabel("K")
plt.ylabel("Inertia score (SSE)")
```

```python
plt.figure(figsize=(15,5))
plt.plot(
    k1[1:],
    ss_core1[1:],
    color="blue",
    linestyle="dashed",
    marker="o",
    markerfacecolor="red",
    markersize=10
)
plt.title("SS vs. K Value")
plt.xlabel("K")
plt.ylabel("SS")
```

```python
kmeans = KMeans(
    n_clusters=5,
    random_state=RANDOM_STATE,
    max_iter=100
).fit(df_clust)
ss_core = ss(df_clust,kmeans.labels_)
inertia= kmeans.inertia_

print("ss_core", ss_core)
print("inertia", inertia)
```

```python
df_features = dataframe.copy()
df_features["cluster_ids"] = kmeans.labels_
df_features = df_features.reset_index()
```

```python
df_profile_overall = df_features.describe().T
df_profile_overall["Overall Dataset"] = df_profile_overall.apply(
    lambda row: row["50%"] if row.name in ["y"] else
    row["count"] if row.name in ["student_id"] else row["mean"],
    axis=1
)
df_profile_overall = df_profile_overall[["Overall Dataset"]]
```

```python
df_cluster_summary = df_features.groupby("cluster_ids").describe().T.reset_index()
df_cluster_summary = df_cluster_summary.rename(
    columns={
        "level_0": "column",
        'level_1': "metric"
    }
)
```

```python
df_cluster_summary = df_cluster_summary.query(
    """
        (column.isin(["y"]) & metric == "50%") | \
        (column.isin(["student_id"]) & metric == "count") | \
        (~column.isin(["student_id", "y"]) & metric == "mean")
    """,
    engine="python")
```

```python
df_cluster_summary["Metric"] = df_cluster_summary.apply(
    lambda row: "median" if row["metric"]=="50%" else row["metric"],
    axis=1
)
df_cluster_summary = df_cluster_summary.set_index("column")
df_cluster_summary = df_cluster_summary.drop("metric", axis=1)
```

```python
df_profile = df_cluster_summary.join(df_profile_overall)
```

```python
df_profile.to_csv("./kmeans_profiling.csv")
```
